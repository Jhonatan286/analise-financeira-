# comparando carteiras com ibov


```python
ativos = ["MELK3.SA", "LAVV3.SA", "MTRE3.SA", "DIRR3.SA", "JHSF3.SA", "CALI3.SA", "MRVE3.SA"]
```

```python
inicio = '2018-01-01'
fim = '2023-01-01'
```

```python
precos = yf.download(ativos, start = inicio, end = fim)['Adj Close']
```

```python
precos.head()
```

```python
precos.columns
```

```python
compras = {'CALI3.SA': 1000, 'DIRR3.SA': 1000, 'JHSF3.SA': 1000, 'LAVV3.SA': 1000, 'MELK3.SA': 1000, 'MRVE3.SA': 1000,
       'MTRE3.SA': 1000}
```

```python
sum(compras.values())
```

```python
primeiro = precos.iloc[0]
```

```python
list(compras.keys())
```

```python
compras_df = pd.Series(data = compras, index = list(compras.keys()))
```

```python
qtd_acoes = compras_df/primeiro
```

```python
PL = precos*qtd_acoes
```

```python
PL.head()
```

```python
PL.tail()
```

```python
PL['PL Total'] = PL.sum(axis = 1)
```

```python
ibov = yf.download('^BVSP', start = inicio, end =  fim)
```

```python
ibov.rename(columns = {'Adj Close': 'IBOV'}, inplace = True)
```

```python
ibov.drop(ibov.columns[[0,1,2,3,5]], axis = 1, inplace = True)
```

```python
ibov.head()
```

```python
consolidado = pd.merge(ibov, PL, how = 'inner', on = 'Date')
```

```python
consolidado.head()
```

```python
consolidado.plot()
```

```python
consolidado_adj = consolidado/consolidado.iloc[0]
```

```python
consolidado_adj.head()
```

```python
consolidado_adj.tail()
```

```python
consolidado_adj.plot()
```

```python
consolidado_adj[['IBOV', 'PL Total']].plot(figsize = (15,10));
```
